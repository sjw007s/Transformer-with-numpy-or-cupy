```python


def gan_forward_postproc(self, output, y):
    #entropy = sigmoid_cross_entropy_with_logits(y, output)
    #loss = np.mean(entropy)
    #aux = [y, output]
    
    
    #print("output",output)
    #print("y",y)
    diff = (output - y).astype(np.float32)
    #print("diff",diff)
    square = np.square(diff)
    #print("square",square)
    loss = np.mean(square).astype(np.float32)
    
    #print("loss",loss)
    return loss, diff

def gan_backprop_postproc(self, G_loss, diff):
    #y, output = aux
    #shape = output.shape
    
    #g_loss_entropy = np.ones(shape) / np_cpu.prod(shape)
    #g_entropy_output = sigmoid_cross_entropy_with_logits_derv(y, output)

    #G_entropy = g_loss_entropy * G_loss
    #G_output = g_entropy_output * G_entropy
    
    shape = diff.shape
    
    g_loss_square = np.ones(shape, dtype = 'float32') / np_cpu.prod(shape)
    g_square_diff = 2 * diff
    g_diff_output = 1

    G_square = g_loss_square * G_loss
    G_diff = g_square_diff * G_square
    G_output = g_diff_output * G_diff
    
    return G_output

GanDataset.forward_postproc = gan_forward_postproc
GanDataset.backprop_postproc = gan_backprop_postproc
```
